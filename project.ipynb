```python
import pandas as pd


ds = pd.read_csv("./assets/real_estate.csv", delimiter=';')


valdemorillo_df = ds[ds['level5'] == 'Valdemorillo']
galapagar_df = ds[ds['level5'] == 'Galapagar']

valdemorillo_df_cleaned = valdemorillo_df.dropna(subset=['price', 'surface'])
valdemorillo_df_cleaned = valdemorillo_df_cleaned[(valdemorillo_df_cleaned['price'] != 0)
                         & (valdemorillo_df_cleaned['surface'] != 0)]

#print(valdemorillo_df_cleaned)

galapagar_df_cleaned = galapagar_df.dropna(subset=['price', 'surface'])
galapagar_df_cleaned = galapagar_df_cleaned[(galapagar_df_cleaned['price'] != 0) 
                        & (galapagar_df_cleaned['surface'] != 0)]
#print(galapagar_df_cleaned)
valdemorillo_df_cleaned['pps'] = valdemorillo_df_cleaned['price'] / valdemorillo_df_cleaned['surface']
galapagar_df_cleaned['pps'] = galapagar_df_cleaned['price'] / galapagar_df_cleaned['surface']

avg_pps_valdemorillo = valdemorillo_df_cleaned['pps'].mean()
avg_pps_galapagar = galapagar_df_cleaned['pps'].mean()

print("Avg. Price/m2 Valdemorillo:", avg_pps_valdemorillo)
print("Avg. Price/m2 Galapagar:   ", avg_pps_galapagar)
```

#### Exercise 16. Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (you must use the subset obtained in the question 14) (★★★)

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv("./assets/real_estate.csv", delimiter=';')

south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

south_belt_subset = ds[ds['level5'].isin(south_belt_populations)].dropna(subset=['price'])

normalized_prices = {}
for population in south_belt_populations:
    prices = south_belt_subset[south_belt_subset['level5'] == population]['price']
    min_price = prices.min()
    max_price = prices.max()
    normalized_prices[population] = (prices - min_price) / (max_price - min_price)

plt.figure(figsize=(10, 6))
for population in south_belt_populations:
    plt.hist(normalized_prices[population], bins=20, alpha=0.5, label=population)

plt.title('Normalized Prices Distribution by Population')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

##### Comment
* Se peude apreciar como Getafe destaca sobre el resto de zonas al tener un numero similar de viviendas cerca del precio minimo y con valor del 40% superior.

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
import pandas as pd
import matplotlib.pyplot as plt


ds = pd.read_csv("./assets/real_estate.csv", delimiter=';')

south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
south_belt_subset = ds[ds['level5'].isin(south_belt_populations)].dropna(subset=['price'])

# Create subplots for each population
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Iterate over each population and plot its scatter plot
for idx, population in enumerate(south_belt_populations):
       #wanna try it with greater indexes
        row = idx // 2
        col = idx % 2
        ax = axes[row, col]

        # Filter data for the current population
        population_data = south_belt_subset[south_belt_subset['level5'] == population]

        # Plot scatter plot
        ax.scatter(population_data['surface'], population_data['price'], alpha=0.3)
        ax.set_title(population)
        ax.set_xlabel('Surface')
        ax.set_ylabel('Price')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from ipyleaflet import Map, basemaps, CircleMarker

ds = pd.read_csv("./assets/real_estate.csv", delimiter=';')

south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
south_belt_subset = ds[ds['level5'].isin(south_belt_populations)]


# Create a map centered on the first coordinate in the south_belt_subset
first_lat, first_lon = south_belt_subset.iloc[0]['latitude'], south_belt_subset.iloc[0]['longitude']
map = Map(center=(first_lat, first_lon), zoom=10)

# Define colors for each population
population_colors = {
    "Fuenlabrada": "blue",
    "Leganés": "green",
    "Getafe": "red",
    "Alcorcón": "orange"
}

# Plot coordinates of each population on the map
for population, color in population_colors.items():
    population_data = south_belt_subset[south_belt_subset['level5'] == population]
    for _, row in population_data.iterrows():
        latitude, longitude = row['latitude'], row['longitude']
        marker = CircleMarker(location=(latitude, longitude), color=color, radius=5)
        map.add_layer(marker)

map
```
